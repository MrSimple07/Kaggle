```python
import pandas as pd
df = pd.read_csv('/kaggle/input/world-population-data/world_population_data.csv')
df.head()
```

```python
import matplotlib.pyplot as plt

countries_of_interest = ['India', 'China', 'United States', 'Uzbekistan']
selected_years = ['2023 population', '2015 population', '2000 population', '1980 population', '1970 population']

plt.figure(figsize=(12, 6))
for country in countries_of_interest:
    plt.plot(selected_years, df[df['country'] == country][selected_years].values.flatten(), label=country, marker='o')

plt.title('Population Time Series for Selected Countries')
plt.xlabel('Year')
plt.ylabel('Population')
plt.legend()
plt.grid(True)
plt.show()
```

As we can see, the population of China and India started growing from 1980 year.
