**Population in different continents and their connection between happiest countries**

```python
import plotly_express as px
import matplotlib.pyplot as plt
import seaborn as sns

df_sum = df.groupby('region')['population_2024'].sum().reset_index(name='Population')

fig = px.bar(df_sum, x='region', y='Population', title='Total Population by Region')

fig.show()
```

**Correlation analysis between population density, growth rate, land area and population**

```python
df_corr = ['population_2024', 'land_area', 'population_growthRate', 'population_densityMi', 'WorldHappiness2022']

sns.heatmap(df[df_corr].corr(), cmap ='viridis', annot = True)
plt.show()
```

**Regional Population Density Comparison**

```python
import plotly.express as px

fig_bar = px.bar(df.groupby('region')['population_density'].mean().reset_index(),
                 x='region',
                 y='population_density',
                 title='Regional Population Density Comparison',
                 labels={'population_density': 'Population Density'},
                 color='region')
fig_bar.show()
```

**The correlation between Human Development Index (HDI) and population density **

```python
import plotly.express as px

fig_scatter = px.scatter(df, x='Hdi2021', y='population_density',
                         title='HDI vs. Population Density',
                         labels={'Hdi2021': 'Human Development Index (HDI)', 'population_density': 'Population Density'},
                         hover_name='country')

fig_heatmap = px.density_heatmap(df, x='Hdi2021', y='population_density',
                                 title='HDI vs. Population Density Heatmap',
                                 labels={'Hdi2021': 'Human Development Index (HDI)', 'population_density': 'Population Density'})

fig_scatter.show()
fig_heatmap.show()
```

```python
df.head()
```

**Land Area vs. Population Density Scatter Plot:**

```python
import plotly.express as px

fig_scatter = px.scatter(df, x='land_area', y='population_density',
                         title='Land Area vs. Population Density',
                         labels={'land_area': 'Land Area', 'population_density': 'Population Density'},
                         hover_name='country')

fig_scatter.update_layout(xaxis_type="log", yaxis_type="log")  # Logarithmic scale for better visualization

fig_scatter.show()
```

**PMap visualization**

```python
fig = px.choropleth(df.groupby('country')['population_2024'].sum().reset_index(), 
                    locations='country',
                    locationmode='country names',
                    color= 'population_2024',
                    hover_name='country',
                    color_continuous_scale='Rainbow')
fig.update_geos(projection_type="natural earth", showcoastlines=True)
fig.update_layout(title_text='World Map - population')
fig.show()
```

**Top 5 best countries to live**

```python
import plotly.express as px

top_countries = df.nlargest(5, 'WorldHappiness2022')

fig = px.bar(top_countries, x= 'country', y = 'WorldHappiness2022')
fig.update_layout(xaxis_title = 'Country', yaxis_title = 'Happiness report')
fig.show()
```

**Best countries to live Map**

```python
import plotly.express as px

fig = px.choropleth(df.groupby('country')['WorldHappiness2022'].sum().reset_index(),
                    locations= 'country',
                    locationmode = 'country names',
                    color = 'WorldHappiness2022',
                    hover_name = 'country',
                    color_continuous_scale = 'Rainbow'
                   )
fig.update_geos(projection_type = 'natural earth', showcoastlines = True)
fig.update_layout(title_text = 'World Map - Happiest countries')
fig.show()
```
