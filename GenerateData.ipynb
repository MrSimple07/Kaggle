# Student Exam Prediction Data

```python
import pandas as pd
import numpy as np

np.random.seed(42)

data = {
    'Study Hours': np.random.uniform(1, 10, size=500),
    'Previous Exam Score': np.random.uniform(40, 100, size=500),
    'Pass/Fail': np.random.choice([0, 1], size=500)
}

df = pd.DataFrame(data)
df['Pass/Fail'] = df.apply(lambda row: 1 if (row['Study Hours'] > 5 and row['Previous Exam Score'] > 60) else 0, axis=1)

df.to_csv('student_exam_data_new.csv', index=False)
```

```python
import pandas as pd
import numpy as np

np.random.seed(42)

# Number of samples
num_samples = 1000

# Generate random data for the features
number_of_customers = np.random.randint(10, 100, num_samples)
menu_price = np.random.uniform(10, 50, num_samples)
marketing_spend = np.random.uniform(0, 20, num_samples)

# Generate revenue based on a simple formula (randomized for variety)
revenue = 3 * number_of_customers + 2 * menu_price + 5 * marketing_spend + np.random.normal(0, 30, num_samples)

# Generate random data for the new columns
cuisine_types = np.random.choice(['Italian', 'Mexican', 'Japanese', 'American'], num_samples)
average_customer_spending = np.random.uniform(10, 50, num_samples)
promotions = np.random.randint(0, 2, num_samples)  # Binary: 0 or 1
reviews = np.random.randint(0, 100, num_samples)
monthly_revenue = revenue + np.random.normal(0, 50, num_samples)

# Create a DataFrame
restaurant_data = pd.DataFrame({
    'Number_of_Customers': number_of_customers,
    'Menu_Price': menu_price,
    'Marketing_Spend': marketing_spend,
    'Revenue': revenue
})

restaurant_data['Cuisine_Type'] = cuisine_types
restaurant_data['Average_Customer_Spending'] = average_customer_spending
restaurant_data['Promotions'] = promotions
restaurant_data['Reviews'] = reviews
restaurant_data['Monthly_Revenue'] = monthly_revenue

print(restaurant_data.head())
restaurant_data.to_csv('restaurants_revenue.csv', index=False)

```

```python
import numpy as np
import pandas as pd
from datetime import datetime, timedelta

np.random.seed(14)

num_points = 1000

start_date = datetime(2022, 1, 1, 0, 0, 0)
timestamps = [start_date + timedelta(hours=i) for i in range(num_points)]

# Generate random values for features
temperature = np.random.uniform(20, 30, num_points)
humidity = np.random.uniform(30, 60, num_points)
square_footage = np.random.uniform(1000, 2000, num_points)
occupancy = np.random.randint(0, 10, num_points)
hvac_usage = np.random.choice(['On', 'Off'], num_points)
lighting_usage = np.random.choice(['On', 'Off'], num_points)
renewable_energy = np.random.uniform(0, 30, num_points)
day_of_week = np.random.choice(['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'], num_points)
holiday = np.random.choice(['Yes', 'No'], num_points)

# Calculate energy consumption (simplified formula)
energy_consumption = 20 + 2 * temperature + 0.5 * occupancy + 5 * (hvac_usage == 'On') + 2 * (lighting_usage == 'On') + 0.1 * renewable_energy + np.random.normal(0, 5, num_points)

# Create a DataFrame
data = pd.DataFrame({
    'Timestamp': timestamps,
    'Temperature': temperature,
    'Humidity': humidity,
    'SquareFootage': square_footage,
    'Occupancy': occupancy,
    'HVACUsage': hvac_usage,
    'LightingUsage': lighting_usage,
    'RenewableEnergy': renewable_energy,
    'DayOfWeek': day_of_week,
    'Holiday': holiday,
    'EnergyConsumption': energy_consumption
})

print(data.head())

```

```python
data.to_csv('Energy_consumption.csv', index = False)
```

# Obesity data Prediction

```python
import numpy as np
import pandas as pd

np.random.seed(42)


num_samples = 1000


ages = np.random.randint(18, 80, num_samples)
genders = np.random.choice(['Male', 'Female'], num_samples)
heights = np.random.normal(loc=170, scale=10, size=num_samples)
weights = np.random.normal(loc=70, scale=15, size=num_samples)
physical_activity_level = np.random.randint(1, 5, num_samples)


bmi = weights / ((heights / 100) ** 2)


obesity_categories = []
for val in bmi:
    if val < 18.5:
        obesity_categories.append('Underweight')
    elif 18.5 <= val < 25:
        obesity_categories.append('Normal weight')
    elif 25 <= val < 30:
        obesity_categories.append('Overweight')
    else:
        obesity_categories.append('Obese')


data = pd.DataFrame({
    'Age': ages,
    'Gender': genders,
    'Height': heights,
    'Weight': weights,
    'BMI': bmi,
    'PhysicalActivityLevel': physical_activity_level,
    'ObesityCategory': obesity_categories
})

data.to_csv('obesity_data.csv', index=False)
```

```python
import numpy as np
import pandas as pd

# Set random seed for reproducibility
np.random.seed(42)

# Generate synthetic weather data
num_days = 365  # Number of days for which we want to generate data
num_locations = 10  # Number of weather stations/locations

# Generate dates for the dataset
dates = pd.date_range(start='2023-01-01', periods=num_days)

# Generate random temperature data for each location
mean_temperatures = np.random.uniform(10, 30, size=num_locations)  # Mean temperature range: 10°C to 30°C
temperature_variability = np.random.uniform(0, 5, size=num_locations)  # Temperature variability
temperatures = np.random.normal(mean_temperatures, temperature_variability, size=(num_days, num_locations))

precipitation = np.random.uniform(0, 20, size=(num_days, num_locations))

wind_speeds = np.random.uniform(0, 15, size=(num_days, num_locations))

# Create a DataFrame to store the synthetic weather data
weather_data = pd.DataFrame({
    'Date': np.repeat(dates, num_locations),
    'Location': np.tile(np.arange(num_locations), num_days),
    'Temperature (C)': temperatures.flatten(),
    'Precipitation (mm)': precipitation.flatten(),
    'Wind Speed (m/s)': wind_speeds.flatten()
})

weather_data.to_csv('weather_data.csv', index=False)

print("Synthetic Weather Data:")
print(weather_data.head())
```

```python
import numpy as np
import pandas as pd

# Set random seed for reproducibility
np.random.seed(0)

# Number of samples
num_samples = 1000

# Generate random values for features
education_levels = ['High School', 'Bachelor', 'Master', 'PhD']
education = np.random.choice(education_levels, size=num_samples)

experience = np.random.randint(1, 30, size=num_samples)
location = np.random.choice(['Urban', 'Suburban', 'Rural'], size=num_samples)
job_title = np.random.choice(['Engineer', 'Manager', 'Analyst', 'Director'], size=num_samples)
age = np.random.randint(20, 65, size=num_samples)
gender = np.random.choice(['Male', 'Female'], size=num_samples)

# Generate salary based on features
base_salary = {
    'High School': 40000,
    'Bachelor': 60000,
    'Master': 80000,
    'PhD': 100000
}

# Adjust salary based on other factors
salary = []
for edu, exp, loc, job in zip(education, experience, location, job_title):
    base = base_salary[edu]
    exp_bonus = exp * 1000  # Assuming $1000 per year of experience
    loc_bonus = {'Urban': 10000, 'Suburban': 5000, 'Rural': 0}[loc]
    job_bonus = {'Engineer': 10000, 'Manager': 20000, 'Analyst': 5000, 'Director': 30000}[job]
    total_salary = base + exp_bonus + loc_bonus + job_bonus
    # Adding some random noise
    total_salary += np.random.normal(0, 10000)
    salary.append(max(0, total_salary))  # Ensure salary is non-negative

# Create DataFrame
data = pd.DataFrame({
    'Education': education,
    'Experience': experience,
    'Location': location,
    'Job_Title': job_title,
    'Age': age,
    'Gender': gender,
    'Salary': salary
})

# Save to CSV
data.to_csv('salary_prediction_data.csv', index=False)
```

# Car prices Prediction

```python
import numpy as np
import pandas as pd

num_samples = 1000

makes = ['Toyota', 'Honda', 'Ford', 'Chevrolet', 'Nissan']
models = ['Camry', 'Civic', 'F-150', 'Silverado', 'Altima']
years = np.random.randint(2010, 2023, size=num_samples)
mileages = np.random.randint(10000, 150000, size=num_samples)
conditions = ['Excellent', 'Good', 'Fair']
conditions_prob = [0.6, 0.3, 0.1]

def generate_car_prices(num_samples):
    prices = []
    for i in range(num_samples):
        base_price = 20000
        year_factor = (2022 - years[i]) * 1000
        mileage_factor = -0.05 * mileages[i]
        condition_factor = np.random.choice(conditions, p=conditions_prob)
        if condition_factor == 'Excellent':
            condition_price_factor = 0.1
        elif condition_factor == 'Good':
            condition_price_factor = 0
        else:
            condition_price_factor = -0.1
        prices.append(base_price + year_factor + mileage_factor + condition_price_factor)
    return prices

data = {
    'Make': np.random.choice(makes, size=num_samples),
    'Model': np.random.choice(models, size=num_samples),
    'Year': years,
    'Mileage': mileages,
    'Condition': np.random.choice(conditions, p=conditions_prob, size=num_samples),
    'Price': generate_car_prices(num_samples)
}

df = pd.DataFrame(data)

df.to_csv('CarPricesPrediction.csv')

print(df.head())
```

# Diabetes Prediction

```python
import numpy as np
import pandas as pd

np.random.seed(42)

features = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age']

feature_distributions = {
    'Pregnancies': np.random.poisson(1.8, 1000),
    'Glucose': np.random.normal(100, 20, 1000),
    'BloodPressure': np.random.normal(72, 14, 1000),
    'SkinThickness': np.random.normal(23.3, 1.2, 1000),
    'Insulin': np.random.normal(88.3, 70.8, 1000),
    'BMI': np.random.normal(25.6, 3.6, 1000),
    'DiabetesPedigreeFunction': np.random.uniform(0.1, 0.8, 1000),
    'Age': np.random.normal(43, 14, 1000)
}

diagnosis = np.where(
    (np.random.rand(1000) < 0.3) |
    ((np.random.rand(1000) < 0.01) &
    np.logical_or.reduce([feature_distributions[feature] > threshold for feature, threshold in
    zip(['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI', 'DiabetesPedigreeFunction'],
    [5, 130, 130, 35, 100, 30, 0.45])])), 1, 0)

df = pd.DataFrame(dict(zip(features, [list(feature_distributions[feature]) for feature in features])))
df['Diagnosis'] = diagnosis

df = df.sample(frac=1).reset_index(drop=True)

df.to_csv('Diabetes_prediction.csv', index=False)
```
